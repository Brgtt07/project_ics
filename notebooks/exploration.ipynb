```python
import pandas as pd
df = pd.read_csv("../raw_data/merged_country_level/final_merged_dataset_with_knn.csv")
df.describe()
```

```python
# Define a DataFrame for numeric features
df_numeric = df.select_dtypes(include=['float64', 'int64'])

# Calculate median values
median_values = df_numeric.median()
print("\n--- Median Values ---")
print(median_values)

```

```python

# Additional exploration: Show correlations between numerical columns
correlation_matrix = df_numeric.corr()
print("\n--- Correlation Matrix ---")
correlation_matrix

# Visualize the correlation matrix using a heatmap (if matplotlib is available)
try:
    import matplotlib.pyplot as plt
    import seaborn as sns

    plt.figure(figsize=(10, 8))
    sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True)
    plt.title("Correlation Matrix Heatmap")
    plt.show()
except ImportError:
    print("Matplotlib or Seaborn not available for visualization.")
```
