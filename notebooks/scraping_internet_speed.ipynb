```python
soup = BeautifulSoup(response.text, "html.parser")
```

```python
table = soup.find("table", class_="list-results")
```

```python

# Check if the request was successful
if response.status_code == 200:
    soup = BeautifulSoup(response.text, "html.parser")

    # Find the table with the class "list-results"
    table = soup.find("table", class_="list-results")

    # Extract table headers
    headers = [th.text.strip() for th in table.find("thead").find_all("th")]

    # Extract table rows
    rows = []
    for tr in table.find("tbody").find_all("tr"):
        cells = [td.text.strip() for td in tr.find_all("td")]
        rows.append(cells)

    # Convert to DataFrame
    df = pd.DataFrame(rows, columns=headers)

    # Save to CSV
    df.to_csv("internet_speed_rankings.csv", index=False)

    print("Data successfully scraped and saved to internet_speed_rankings.csv")
else:
    print(f"Failed to retrieve the webpage, status code: {response.status_code}")
```

### Now we parse the element's data and put into a CSV

```python
soup = BeautifulSoup(element, "html.parser")

# Find the table
table = soup.find("table", class_="list-results")

# Extract rows
data = []
for row in table.find_all("tr", class_="data-result results"):
    # Extract country name (inside <td class="country">, with an <a> tag)
    country_td = row.find("td", class_="country")
    country = country_td.find("a").text.strip() if country_td.find("a") else country_td.text.strip()
    
    # Extract internet speed (inside <td class="speed">)
    speed = row.find("td", class_="speed").text.strip()

    # Append to data list
    data.append([country, speed])

# Convert to DataFrame
df = pd.DataFrame(data, columns=["Country", "Internet Speed (Mbps)"])

# Save to CSV
df.to_csv("internet_speed_rankings.csv", index=False)

print("Data successfully saved to 'internet_speed_rankings.csv'")
```
